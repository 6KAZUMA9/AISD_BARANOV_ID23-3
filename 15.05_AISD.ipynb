Задача 35.4
Создайте двусвязный список для хранения информации о заказах в ресторане. Каждый элемент списка должен содержать название блюда, цену, количество, дату заказа и статус заказа (например, «в обработке», «готов», «доставлен»).

```python
class Order:
    def __init__(self, dish_name, price, quantity, date, status):
        self.dish_name = dish_name
        self.price = price
        self.quantity = quantity
        self.date = date
        self.status = status
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def add_order(self, dish_name, price, quantity, date, status):
        new_order = Order(dish_name, price, quantity, date, status)
        if self.head is None:
            self.head = new_order
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_order
            new_order.prev = current

    def display_orders(self):
        current = self.head
        while current:
            print("Dish Name:", current.dish_name)
            print("Price:", current.price)
            print("Quantity:", current.quantity)
            print("Date:", current.date)
            print("Status:", current.status)
            print("----------------------")
            current = current.next

# Пример использования
orders_list = DoublyLinkedList()
orders_list.add_order("Pizza", 10.99, 2, "2022-01-15", "в обработке")
orders_list.add_order("Pasta", 8.99, 1, "2022-01-16", "готов")

orders_list.display_orders()
```

Задача 36.4
Реализовать функцию, которая удаляет все повторяющиеся элементы из двусвязного списка.

```python
class Order:
    def __init__(self, dish_name, price, quantity, date, status):
        self.dish_name = dish_name
        self.price = price
        self.quantity = quantity
        self.date = date
        self.status = status
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def add_order(self, dish_name, price, quantity, date, status):
        new_order = Order(dish_name, price, quantity, date, status)
        if self.head is None:
            self.head = new_order
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_order
            new_order.prev = current

    def display_orders(self):
        current = self.head
        while current:
            print("Dish Name:", current.dish_name)
            print("Price:", current.price)
            print("Quantity:", current.quantity)
            print("Date:", current.date)
            print("Status:", current.status)
            print("----------------------")
            current = current.next
            
    def remove_duplicates(self):
        current = self.head
        seen_values = set()
        while current:
            if current.dish_name in seen_values:
                if current.next is not None:
                    current.next.prev = current.prev
                if current.prev is not None:
                    current.prev.next = current.next
                else:
                    self.head = current.next
            else:
                seen_values.add(current.dish_name)
            current = current.next

# Пример использования
orders_list = DoublyLinkedList()
orders_list.add_order("Pizza", 10.99, 2, "2022-01-15", "в обработке")
orders_list.add_order("Pasta", 8.99, 1, "2022-01-16", "готов")
orders_list.remove_duplicates()
orders_list.display_orders() 

orders_list.display_orders()
```

Задание 37.4
Реализовать функцию, которая удаляет все элементы циклического двусвязного списка, равные заданному значению

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
            self.head.prev = self.head
        else:
            last = self.head.prev
            last.next = new_node
            new_node.prev = last
            new_node.next = self.head
            self.head.prev = new_node

    def display_list(self):
        if not self.head:
            return
        current = self.head
        while True:
            print(current.data, end=" ")
            current = current.next
            if current == self.head:
                break
        print()

    def remove_elements(self, value):
        if not self.head:
            return

        current = self.head
        while True:
            if current.data == value:
                if current.next == current:  # Если у нас только один элемент в списке
                    self.head = None
                else:
                    current.prev.next = current.next
                    current.next.prev = current.prev
                    if current == self.head:
                        self.head = current.next
            current = current.next
            if current == self.head:
                break

# Пример использования
cdll = CircularDoublyLinkedList()
cdll.append(1)
cdll.append(2)
cdll.append(3)
cdll.append(2)
cdll.append(4)

print("Исходный список:")
cdll.display_list()

value_to_remove = 2
cdll.remove_elements(value_to_remove)

print(f"Список после удаления всех элементов со значением {value_to_remove}:")
cdll.display_list()
```
